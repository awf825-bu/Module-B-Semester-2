# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import pandas as pd
import numpy as np

# Sample historical data
years_hist = ['2006','2007','2008','2009','2010','2011','2012','2013','2014']
data_hist = [50,51,60,61,77,81,83,100,108]

plt.plot(years_hist, data_hist, label='Historical Data', color='blue')

# Sample forecast data
years_forecast = ['2014','2015','2016','2017','2018']
data_forecast = [108,119,131,144,158]

plt.plot(years_forecast, data_forecast, label='Forecast', color='blue', linestyle='--', marker='o')

for i, val in enumerate(data_forecast):
    plt.annotate(f'${val}', (years_forecast[i], data_forecast[i]), textcoords="offset points", xytext=(0,10), ha='center', color='b')

# plt.xlabel('Date')
plt.ylabel('Sales ($Billion)')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
formatter = mticker.StrMethodFormatter('${x}')
plt.gca().yaxis.set_major_formatter(formatter)
plt.gca().text(4, -25, 'ACTUAL', ha='center', va='top')
plt.gca().text(10, -25, 'FORECAST', ha='center', va='top')
plt.gca().axvspan(8, 12, alpha=0.3, color='gray')
plt.yticks([0,20,40,60,80,100,120,140,160,180])
plt.title('Sales Over Time')

plt.tight_layout()
plt.ylim(ymin=0, ymax=180)
plt.show()
```
